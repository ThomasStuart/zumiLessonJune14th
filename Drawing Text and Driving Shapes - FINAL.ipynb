```python
# if left sensor detects an object, zumi drives in a sqaure 
# else if right sensor detects an object, zumi drives in a s shaped 
from zumi.zumi import Zumi
from zumi.util.screen import Screen
import time
screen=Screen()
zumi = Zumi()
zumi.reset_gyro()


# code to turn this into a function 
def drive_square():
    zumi.turn_right()
    zumi.forward(POWER, TIME)
    zumi.turn_left()
    zumi.forward(POWER, TIME*2)
    zumi.turn_left()
    zumi.forward(POWER, TIME*2)
    zumi.turn_left()
    zumi.forward(POWER, TIME*2)
    zumi.turn_left()
    zumi.forward(POWER, TIME)
    
def drive_s():
    POWER = 40 
    angle = 0
    duration = 15

    # first half of the S, going left
    for i in range(0, duration):
        angle = angle + 5  
        zumi.go_straight(POWER ,angle)
        print( angle)

    zumi.go_straight(POWER ,0)

    duration = 25

    # second half of s, going right 
    for i in range(0, duration):
        angle = angle - 5  
        zumi.go_straight(POWER ,angle)
        print("back", angle)

try:
    zumi.reset_gyro()
    ir_readings  = zumi.get_all_IR_data()
    front_left    = ir_readings[5]
    front_right   = ir_readings[0]
    detect_value_left  = 50 
    detect_value_right = 150


    message = "IR_L = square ,  IR_R = S "

    screen.draw_text( message, 0,0)
    time.sleep(2)

    strLeft  = "left"
    strRight = "right"

    x = 125/2
    y = 50/2

    print("left:  ", front_left)
    print("right: ", front_right)
    screen.clear_display()


    while True:

        ir_readings  = zumi.get_all_IR_data()
        front_left    = ir_readings[5]
        front_right   = ir_readings[0]

        # if the left sensor is detecting an object 
        if  front_left < detect_value_left:
            # print left on the zumi screen
            screen.draw_text( strLeft, x, y)
            drive_square()

        # else if the right sensor is detecting an object
        elif front_right < detect_value_right: 
            # print right on zumi screen
            screen.draw_text( strRight, x, y)
            drive_s()

finally:
    zumi.stop()
    print("done")
```
